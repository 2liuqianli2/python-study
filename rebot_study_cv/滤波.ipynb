```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def show(img):
    if img.ndim==2:
        plt.imshow(img,cmap="gray")
    else:
        plt.imshow(cv.cvtColor(img,cv.COLOR_RGB2BGR))
    
```

```python
img=cv.imread("./pic/rose_spnoise_200x200.jpg")
# print(img)
show(img)
```

# 均值滤波

```python
k=np.ones((3,3))/9
```

```python
img2=cv.filter2D(img,-1,k)
show(np.hstack([img,img2]))
```

```python
img3=cv.blur(img,(3,3))
img4=cv.boxFilter(img,-1,(3,3))
show(np.hstack([img3,img4]))
```

# 中值滤波

```python
img5=cv.medianBlur(img,3)
show(np.hstack([img,img5]))
```

# 高斯滤波

```python
img6=cv.GaussianBlur(img,(5,5),sigmaX=0.8)
show(img6)
```

```python
img7=cv.bilateralFilter(img,-1,sigmaColor=50,sigmaSpace=3)
show(img7)
```
