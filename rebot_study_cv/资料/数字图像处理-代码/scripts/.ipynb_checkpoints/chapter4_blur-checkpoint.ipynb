```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.show()
```

## 4 模糊化

```python
img = cv2.imread('pic/rose200x200.jpg', 0)
show(img)
```

```python
hstack = lambda imgs: np.hstack(imgs)
```

### 4.1 均值模糊

```python
img_mblur = cv2.blur(img, (5,5)) #, borderType=cv2.BORDER_CONSTANT
show(np.hstack([img, img_mblur]))
```

```python
img = cv2.imread('pic/rose_gaussiannoise_200x200.jpg', 0)
img_mblur = cv2.blur(img, (5,5)) #, borderType=cv2.BORDER_CONSTANT
show(np.hstack([img, img_mblur]))
```

```python
img = cv2.imread('pic/rose_spnoise_200x200.jpg', 0)
img_mblur = cv2.blur(img, (5,5)) #, borderType=cv2.BORDER_CONSTANT
show(np.hstack([img, img_mblur]))
```

### 4.2 Gaussian模糊

```python
img = cv2.imread('pic/rose_spnoise_200x200.jpg', 0)

img_gblur1 = cv2.GaussianBlur(img, (5,5), 0.5)
img_gblur2 = cv2.GaussianBlur(img, (5,5), 10)
show(hstack([img, img_gblur1]))
show(hstack([img, img_gblur2]))
```

### 4.3 中值模糊

```python
img = cv2.imread('pic/rose_spnoise_200x200.jpg', 0)

img_median = cv2.medianBlur(img, 3)
show(hstack([img, img_median]))
```

```python
img = cv2.imread('pic/rose_gaussiannoise_200x200.jpg', 0)

img_median = cv2.medianBlur(img, 3)
show(hstack([img, img_median]))
```

### 4.4 自定义滤波器

```python
img = cv2.imread('pic/rose200x200.jpg', 0)

kernel = np.ones((5,5), dtype=np.float32)
kernel /= kernel.sum()

img_filter = cv2.filter2D(img, -1, kernel)
show(hstack([img, img_filter]))
```

```python
img = cv2.imread('pic/rose200x200.jpg', 0)

kernel = np.array([
    [1, 1, 1],
    [1, -9, 1],
    [1, 1, 1]
], dtype=np.float32)

img_filter = cv2.filter2D(img, -1, kernel)
show(hstack([img, img_filter]))
```

## 添加噪声

```python
#### 1. 椒盐噪声
def salt_pepper_noise(img, ratio=0.02):
    img = img.copy()
    for _ in range(int(img.size * ratio)):
        i = np.random.randint(0, img.shape[0])
        j = np.random.randint(0, img.shape[1])     
        img[i,j] = np.random.randint(0, 100)
    print(img.min())
    return img

img = cv2.imread('pic/rose200x200.jpg', 0)
img_sp = salt_pepper_noise(img)
show(hstack([img, img_sp]))
```

```python
cv2.imwrite('pic/rose_spnoise_200x200.jpg', img_sp)
```

```python
def gaussian_noise(img, mean=0, var=20):
    noise = np.random.normal(mean, var, img.shape)
    img = img.astype(np.float) + noise
    return img
```

```python
img = cv2.imread('pic/rose200x200.jpg', 0)
img_ga = gaussian_noise(img)
show(hstack([img, img_ga]))
```

```python
cv2.imwrite('pic/rose_gaussiannoise_200x200.jpg', img_ga)
```
