```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img, img2=None):
    if img.ndim == 2:
        if img2 is not None:
            plt.imshow(np.hstack([img, img2], cmap='gray'))
        else:
            plt.imshow(img, cmap='gray')  
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if img2 is not None:
            img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
            plt.imshow(np.hstack([img, img2]))
        else:
            plt.imshow(img)
    plt.show()
```

## 8. 金字塔采样

### 8.1 降采样

```python
img = cv2.imread('pic/cute500x500.jpg')
d = cv2.pyrDown(img)
dd = cv2.pyrDown(d)
```

### 8.2 上采样

```python
ddu = cv2.pyrUp(dd) #(125x125) ->(250, 250)
dduu = cv2.pyrUp(ddu) #(250,250) ->(500,500)

show(img, dduu)
```

### 8.4 图像融合 Orapple

```python
# 参考Opencv官方文档

import cv2
import numpy as np

A = cv2.imread("pic/apple.jpg")
B = cv2.imread("pic/orange.jpg")

# 对A进行高斯金字塔下采样
G = A.copy()
gpA = [G]
for i in range(5):
    G = cv2.pyrDown(G)
    gpA.append(G)
    
# 同样，对B进行高斯金字塔下采样
G = B.copy()
gpB = [G]
for i in range(5):
    G = cv2.pyrDown(G)
    gpB.append(G)
    
# 创建Laplacian金字塔
lpA = [gpA[-1]]
for i in range(5, 0, -1):
    GE = cv2.pyrUp(gpA[i])
    L = cv2.subtract(gpA[i-1], GE)
    lpA.append(L)
    
lpB = [gpB[-1]]
for i in range(5, 0, -1):
    GE = cv2.pyrUp(gpB[i])
    L = cv2.subtract(gpB[i-1], GE)
    lpB.append(L)
    
# lpA中每个元素取左半，lpB中每个元素取右半，拼成一张图
LS = []
for la, lb in zip(lpA, lpB):
    rows, cols, dpt = la.shape
    ls = np.hstack((la[:, :cols//2], lb[:, cols//2:]))
    LS.append(ls)

# show(LS[0])
# 重建
combine = LS[0]
for i in range(1, 6):
    combine = cv2.pyrUp(combine)
    combine = cv2.add(combine, LS[i])
    
show(combine)
```
