```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img):
    if img.ndim == 2:
        plt.imshow(img, cmap='gray')
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.show()
```

## 7. 形态学处理

```python
img = cv2.imread("pic/ninty50x50.jpg", 0)
show(img)
_, img_bin = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
show(img_bin)
```

### 7.1 腐蚀

```python
kernel = np.ones((3,3), dtype=np.int)
img_erode = cv2.erode(img_bin, kernel)
show(np.hstack([img_bin, img_erode]))
```

### 7.2 膨胀

```python
kernel = np.ones((3,3), dtype=np.int) #(1,2)
img_dilate = cv2.dilate(img_bin, kernel) # iterations
show(np.hstack([img_bin, img_dilate]))
```

```python
cv2.imwrite('pic/ninty_erode_dilate.jpg', np.hstack([img_bin, img_erode, img_dilate]))
```

### 7.3 结构元的设计

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
print(kernel)

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
print(kernel)
```

### 7.4 开运算

```python
# 先腐蚀后膨胀

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
img_open = cv2.morphologyEx(img_bin, cv2.MORPH_OPEN, kernel)

show(np.hstack([img_bin, img_open]))
```

### 7.5 闭运算

```python
# 先膨胀后腐蚀

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3)) 
img_close = cv2.morphologyEx(img_bin, cv2.MORPH_CLOSE, kernel) 

show(np.hstack([img_bin, img_close]))
```

```python
cv2.imwrite('pic/ninty_open_close.jpg', np.hstack([img_bin, img_open, img_close]))
```
