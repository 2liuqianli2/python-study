```python
img1=cv.imread("./pic/line500x500.jpg",0)
img2=cv.imread("./pic/mask500x500.jpg",0)
img3=cv.imread("./pic/hedgehog500x500.jpg",0)
img4=cv.imread("./pic/hedgehog_noise_500x500.jpg",0)

def show(img):
    if img.ndim==2:
        plt.imshow(img,cmap="gray")
    else:
        plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
        # cv.cvtColor(pic,cv.COLOR_BGR2RGB)
        

show(np.hstack([img1,img2,img3,img4]))
```

```python
a=img1+img3
show(a)
```

```python
a=cv.add(img1*0.5,img3*0.5)
show(a)
```

```python
rabbit=cv.imread("./pic/rabbit500x333.jpg")
show(rabbit)
```

```python
a=cv.resize(rabbit,(600,500))
show(a)
```

```python
a=cv.imread("./pic/rabbit500x333.jpg")
M=np.array([
[1,0,100],
[0,1,50]
],dtype=np.float32)
pic=cv.warpAffine(a,M,(400,530))
show(pic)
```

```python
H=cv.imread("./pic/parthenon500x750.jpg")
show(H)
```

```python
M=np.array([
[210,50],
[610,270],
[650,470],
[150,450]
],dtype=np.float32)
S=np.array([
[150,50],
[650,50],
[650,470],
[150,470]
],dtype=np.float32)
M=cv.getPerspectiveTransform(M,S)
M
```
