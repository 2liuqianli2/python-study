```python
img1=cv.imread("./pic/line500x500.jpg",0)
img2=cv.imread("./pic/mask500x500.jpg",0)
img3=cv.imread("./pic/hedgehog500x500.jpg",0)
img4=cv.imread("./pic/hedgehog_noise_500x500.jpg",0)

def show(img):
    if img.ndim==2:
        plt.imshow(img,cmap="gray")
    else:
        plt.imshow(cv.cvt(img,cv.COLOR_BGR2RGB))

show(np.hstack([img1,img2,img3,img4]))
```
