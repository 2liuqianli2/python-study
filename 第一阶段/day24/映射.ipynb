```python
df=DataFrame(data={
    'name':['刘备','关羽','赵云','诸葛亮','貂蝉'],
    'age':[23,34,56,78,12],
    'salary':[1000000,200,200,400000,30000],
    'address':['三里屯','永州','邵阳','武钢','佛山']
})
df
```

```python
#
df.replace(to_replace=30000,value=20)
```

```python
df.replace(to_replace=200,method='ffill')#向前填充 limit 限制被填充的个数
```

```python
df.replace(to_replace=200,method='ffill',limit=1)
```

```python
#使用正则
df.replace(to_replace=r'^.{3}$',value='孙悟空',regex=True)
```

```python
#使用字典
df.replace(to_replace={
    '佛山':'广东'
})
```

```python
#替换多个值
df.replace(to_replace=[200,30000],value=[334,3001])
```
