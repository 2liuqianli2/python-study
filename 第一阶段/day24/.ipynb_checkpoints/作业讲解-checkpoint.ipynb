```python
df=pd.read_excel('./商品.xlsx',sheet_name='Sheet1')
df
```

```python
df[df['商品价格']>10]
```

```python
dic={}
for name in list(df.商品名称):
    if name not in dic.keys():
        a=df.loc[df.商品名称==name,'商品数量'].sum()
        dic[name]=a
dic
maxs=max(dic.values())
df[df.商品数量==maxs]
```
