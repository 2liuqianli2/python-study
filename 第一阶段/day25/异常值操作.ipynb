### 排序和随机取样
排序   根据操作调整行索引或者列索引的顺序
    take(索引列表)   可以理解成这里的索引是隐式索引
    随机排序
        np.random.permutation 给定一个范围 将指定范围的数据打乱 返回
    随机抽样
        借助take np.random.randint()来随机取样

```python
df1=DataFrame(data=np.random.normal(size=(10000,5)),columns=list('ABCDE'))
df1
```

```python
#默认操作行
df1.take(np.random.permutation(1000))
```

```python
#指定索引
df1.take([2,0,4],axis=1)
```

```python
#随机获取数据
df1.take(np.random.randint(0,1000,size=100))
```

```python
#上面的可能存在重复
#可以用random模块里的 random.sample(range(1000),10) 这样产生的数据不会有重复的
#或者用 df1.sample()  replace参数为True是产生重复的数 默认为False
```

```python
#数据分组
groupby就是根据什么进行分组
groups 是查看分组情况
```

```python
taobao=pd.read_csv("./taobao_data.csv")
taobao
```

```python
obj=taobao.groupby('位置')
obj.groups
```

```python
#也可以根据分组，求指定分组的聚合函数
```

```python
obj.agg("mean")
```

```python
#传递多个聚合函数的操作 
obj.agg(["mean",'sum'])
```

```python
#字典形式
obj.agg({
    '价格':'mean',
    '成交量':'sum'
})
```
