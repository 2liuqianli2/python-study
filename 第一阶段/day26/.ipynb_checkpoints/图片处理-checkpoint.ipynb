```python
import scipy
from scipy.fftpack import fft2,ifft2

import matplotlib.pyplot as plt
%matplotlib inline
#处理图片的
from scipy import ndimage
```

```python
im=plt.imread('./image.jpg')
plt.imshow(im)
```

```python
#移动图片
a=ndimage.shift(im,shift=[100,200,0],mode='mirror')
plt.imshow(a)
```

```python
#缩放图片
#如果不想改变颜色第三个值一定要为1
b=ndimage.zoom(im,zoom=[0.8,1.8,1])
plt.imshow(b)
```
