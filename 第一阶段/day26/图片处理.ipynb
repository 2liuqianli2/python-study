```python
import scipy
from scipy.fftpack import fft2,ifft2

import matplotlib.pyplot as plt
%matplotlib inline
#处理图片的
from scipy import ndimage
```

```python
im=plt.imread('./image.jpg')
plt.imshow(im)
```

```python
#移动图片
a=ndimage.shift(im,shift=[100,200,0],mode='mirror')
plt.imshow(a)
```

```python
#缩放图片
#如果不想改变颜色第三个值一定要为1
b=ndimage.zoom(im,zoom=[0.8,1.8,1])
plt.imshow(b)
```

```python
#高斯滤波
moon=plt.imread('./moonlanding.png')
a=ndimage.gaussian_filter(moon,sigma=2.4)
plt.imshow(a,cmap='gray')
```

```python
#中值滤波
#size 从给定的数据从中取出的数据计算中值进行滤波
a=ndimage.median_filter(moon,size=4)
plt.imshow(a,cmap='gray')
```

```python
#维纳滤波
from scipy.signal import wiener
```

```python
a=wiener(moon,mysize=6)
plt.imshow(a,cmap='gray')
```
