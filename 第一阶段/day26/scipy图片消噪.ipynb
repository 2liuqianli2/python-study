```python
import numpy as np
import pandas as pd
import scipy
#scipy.fftpack 进行傅里叶变换的
from scipy.fftpack import fft2,ifft2 


import matplotlib.pyplot as plt
#在当前文档中显示图片
%matplotlib inline
```

```python
moon=plt.imread('./moonlanding.png')
```

```python
moon
```

```python
plt.imshow(moon,cmap='gray')
```

```python
#获取图片对应的频域
f_moon=fft2(moon)
f_moon
```

```python
#消噪的基准点
level=850

f_moon[f_moon>level]=200
f_moon[f_moon<-level]=-200

#将频域转化为时域
n_moon=ifft2(f_moon)
n_moon
```

```python
#提取实部部分
real=n_moon.real
plt.imshow(real,cmap='gray')
```
