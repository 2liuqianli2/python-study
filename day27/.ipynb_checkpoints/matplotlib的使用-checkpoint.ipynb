```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib as mpl
import matplotlib.pyplot as plt #绘图工具

%matplotlib inline
```

## 绘制一个简单的曲线

```python
x=np.linspace(0,np.pi*2,100)
y=np.sin(x)
plt.plot(y) #可以接受一个值 这个值是纵轴的值 横轴会以x值的隐式索引进行填充
plt.plot(x,y)
```

```python
plt.plot(x,y)
```

```python
#一个plot可以绘制多个图线性图
y1=np.cos(x)
plt.plot(x,y,x,y1)
```

```python
#获取画板
"""
figsize 设置画板的宽和高 单位是英寸 也可以是纵向和横向的缩放程度
dpi 绘制图片的分辨率 默认八十
facecolor 画板的背景颜色
edgecolor 边框颜色
frameon 是否显示边框
"""
plt.figure(figsize=(4,4),facecolor='pink')
plt.plot(x,y)
```

```python
#设置子画布
fi=plt.figure(figsize=(8,8),facecolor='pink')
#添加子画布
nine=fi.add_subplot(3,3,9)
nine.plot(x,y)
nine=fi.add_subplot(3,3,6)
nine.plot(x,y)
nine=fi.add_subplot(3,3,1)
nine.plot(x,y)
nine=fi.add_subplot(3,3,7)
nine.plot(x,y)
```

```python
#另外一种形式设置子画布
figure=plt.figure(figsize=(8,8),facecolor='gray')
s=plt.subplot(3,3,7)
s.plot(x,y)
s=plt.subplot(3,3,1)
s.plot(x,y)
s=plt.subplot(3,3,3)
s.plot(x,y)
s=plt.subplot(3,3,8)
s.plot(x,y)
```

```python
#网格线
"""
lw 线的粗细
color 线的颜色
axis 表示水平线还是垂直线 默认是both 参数x x轴引发的线 参数y y轴引发的线
alpha 设置透明度 0到1之间 0完全透明
"""
plt.grid(lw=3,axis='x',color='cyan',alpha=0.1)
```

```python
#坐标轴界限
plt.grid()
# xmin xmax ymin ymax
plt.axis([0,4,-1,1])
plt.plot(x,y)
```

```python
#单独设置某个轴的
plt.xlim([0,4])
plt.ylim([0,1])
plt.plot(x,y)
```

```python
#设置每个画布的坐标轴界限
ab=plt.figure(figsize=(8,8))
ax1=ab.add_subplot(2,2,2)
ax1.axis([0,2,0,1])
ax1.plot(x,y)

ax1=ab.add_subplot(2,2,4)
ax1.axis([0,3,0,1])
ax1.plot(x,y)

ax1=ab.add_subplot(2,2,1)
ax1.axis([0,6,0,1])
ax1.plot(x,y)
```

```python
#设置坐标轴的类型
plt.plot(x,y)
plt.axis('off') #默认是on 
```

```python
# 设置坐标的标签
#xlable ylable
"""
第一个参数是标签值
loc 设置标签的位置 left center right 默认是center
fontsize 设置字号
color 设置颜色
rotation 旋转 指定角度
"""
plt.xlabel('x_lable',loc='left',color='gray')
plt.plot(x,y)
```

```python
#图例
plt.plot(x,y,label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
#设置图例显示
plt.legend()
#参数loc设置位置
#参数ncol 是设置图例有几列 默认是1列
```

```python
#全局设置字体
#查看所有的全局属性设置
mpl.rc_params()
```

```python
#设置标题
plt.title('余弦波')
plt.plot(x,y1)
```
