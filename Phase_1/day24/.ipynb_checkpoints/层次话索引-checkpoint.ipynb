## DataFrame层次化索引

```python
##拥有层次索引
#行标题
index=['zhx','lxy','wxl','ck','cq']
#列标题
columns=[
    ['First','First','First','Second','Second','Second'],
    ['python','java','c','python','java','c']
]
df=DataFrame(data=np.random.randint(30,100,size=(5,6)),index=index,columns=columns)
df
```

```python
df.columns
```

```python
#显示创建DF
# 构建多索引的方式  构建MultiIndex
# 1，使用元组
#2，使用数组
#3，使用乘积
```

```python
#使用数组创建MultiIndex
columns=[['First','First','First','Second','Second','Second'],
    ['python','java','c','python','java','c']
]
mindex=pd.MultiIndex.from_arrays(columns)
mindex
```

```python
#使用元组创建MultiIndex
columns=[( 'First','python'),
            ( 'First','java'),
            ( 'First','c'),
            ('Second','python'),
            ('Second','java'),
            ('Second','c')
]
mindex=pd.MultiIndex.from_tuples(columns)
mindex
```

```python
#使用乘积
value=[
    ['Frist','Second'],
    ['pyton','java','c']
]
mindex=pd.MultiIndex.from_product(value)
mindex
```

```python
#行层次索引
dt=DataFrame(data=np.random.randint(23,100,size=(6,5)),index=mindex,columns=index)
dt
```

### 层次索引下怎么进行 索引和切片

```python

#行标题
index=['zhx','lxy','wxl','ck','cq']
#列标题
columns=[
    ['First','First','First','Second','Second','Second'],
    ['python','java','c','python','java','c']
]
df=DataFrame(data=np.random.randint(30,100,size=(5,6)),index=index,columns=columns)
df
```

```python
# 访问列
#df[('第一级标题'，'第二季标题')]
df[('First','c')]
df.loc[:,('First','c')]
```

```python
#多列
df[[('First','java'),('Second','c')]]
df.loc[:,[('First','java'),('Second','c')]]
```

```python
#获取行
df.loc['zhx']
```

```python
#访问元素
df.loc["cq",('First','python')]
#修改
df.loc["cq",('First','python')]=100
df
```

```python
#多层列索引 切片
#把多层换为单层 再去切片
df1=df['First']
df1.loc[:,'python':'java']
```

```python
#练习
cl=[
    ['第一学期','第二学期'],
    ['python','java','c']
]
cl=pd.MultiIndex.from_product(cl)
df=DataFrame(data=np.random.randint(30,100,size=(4,6)),index=['jack','lucy','tom','meray'],columns=cl)
df
```

## pandas中的汇总操作
    级联--两个df的合并 数据不会进行累计
        pd.concat
    合并--两个df相同字段数据进行累计
        pd.merge

```python
df1=DataFrame(data=np.random.randint(20,100,size=(3,3)),columns=list("ABC"))
df2=DataFrame(data=np.random.randint(20,100,size=(3,4)),columns=list("ABCD"))
display(df1,df2)
```

```python
pd.concat([df1,df2])#合并行
```

```python
pd.concat([df1,df2],axis=1)#合并列
```

```python
"""
join 
    合并形式  outer 默认的 结果取两表的并集
                inner 结果取两表的交集
    
"""
pd.concat([df1,df2],join='inner')
```

```python
#忽略显示索引 ignore_index
pd.concat([df1,df2],ignore_index=True)
```
